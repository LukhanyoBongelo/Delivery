```python
import matplotlib.pyplot as plt
```

```python
age = new_file["Delivery_person_Age"]
```

```python
plt.hist(age, bins=5, edgecolor = "black")
```

```python
rates = new_file["Delivery_person_Ratings"]
```

```python
plt.hist(rates, bins = 10, edgecolor = "black")
```

```python
plt.hist(age, bins=10, edgecolor="Black")
```

```python
corr = new_file.corr()
```

```python
corr
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f",linewidths=0.5)
plt.title("Correlation heatmap of delivery dataset")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x=age, y=rates, color="blue")
plt.xlabel("Deliver person age")
plt.ylabel("Delivery person ratings")
plt.title("Delivery rating vs age")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.regplot(x=age, y=rates, color="red")
plt.xlabel("Delivery person age")
plt.ylabel("Delivery person ratings")
plt.title("age vs ratings")
plt.show()
```
